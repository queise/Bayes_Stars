### Goals:
Calculate the Bayesin posterior probability distribution on the characteristics of a observed star, namely:
- the mass of the star: $M_{\star}$,
- its age: $t_{\star}$,
- its metallicity: $Z$,

comparing the measurements of its:
- effective temperature: $T_{eff}$,
- surface gravity: $\log g$,
- surface metallicity: $\log(Z/Z_{\odot})$,
- and the measured frequencies of acoustic oscillations of the star, $\nu_{n,l}$ for modes of degree $l$ and radial order $n$,

with the results of our stellar evolution code CESAM2k, the adiabatic oscillations code ADIPLS and the code to compute asteroseismic parameters SepCalc.

This calculation will be performed taking into account the influence of Dark Matter (DM) and without it, in order to assess how the inclusion of DM impacts the determination of the stellar parameters.

### Methods:
We use [MultiNest](http://www.multinest.org), a multimodal nested sampling algorithm, to derive the **Bayesian posterior probability distributions** of $[M_{\star}, t_{\star}, Z]$.

#### Priors:

We use a Salpeter Initial Mass Function as a prior for $M_{\star}$:

$
 \xi (M_{\star}) \Delta (M_{\star}) =\xi_0 \left( \frac{M_{\star}}{M_{\odot}} \right)^{-2.35} \left(\frac{\Delta M_{\star}}{M_{\odot}}\right) \;,
$

and flat priors for $t_{\star}$ and $Z$, within the ranges:

$Z \in [0.0005,0.0400]$

$t_{\star} \in [0,13.8]$ (13.8 Gyr is the age of the Universe)


#### Likelihood:

Multinest calls our other codes to calculate the evidence, with an associated error estimate, and produces posterior samples from distributions that may contain multiple modes and pronounced (curving) degeneracies in high dimensions.

The likelihood is defined as:

$
\mathcal {L}=(\prod^{n}_{i=1}\frac{1}{\sqrt{2\pi}\sigma_{i}})\times
 \exp(-\chi^{2}/2),
$

where

$
 \chi^{2}=\sum^{n}_{i=1}(\frac{q^{obs}_{i}-q^{model}_{i}}{\sigma^{i}})^{2}.
$

As we may include parameters into the fitting which are correlated, like the seismic parameters $\langle\Delta\nu\rangle$, $\langle\delta\nu\rangle$, $r_{010}$, we should then include the covariance matrix in the calculation of the $\chi^2$:

$
\chi^2 =  \left(\vec{q}_{\mathrm{obs}} - \vec{q}_{\rm model}\right)^{T} \, {\bf C}^{-1} \, \left(\vec{q}_{\mathrm{obs}} - \vec{q}_{\rm model}\right) \,,
$

## 1) Calculation of the correlation matrix between seismic parameters

### 1.1) Load and store data

```python
# we load the observed frequencies:
l_n = np.loadtxt('Dushera.freqs', usecols = (0,1), dtype=int) #[('f0',int),('f1',int),('f2',float),('f3',float)])
f_e = np.loadtxt('Dushera.freqs', usecols = (2,3))
# store the radial order (n) and degree (l)
l_max = np.max([row[0] for row in l_n])
n_max = np.zeros(l_max+1, dtype=int) ; n_min = np.zeros(l_max+1, dtype=int)
for l in xrange(l_max+1):
    all_ns_of_l = [n for ls,n in l_n if ls==l]
    n_max[l] = np.max(all_ns_of_l)
    n_min[l] = np.min(all_ns_of_l)
    print ("l=%i \t nmin=%i \t nmax=%i " % (l, n_min[l], n_max[l]))
```

```python
# creating dictionaries for freq[l][n] and err[l][n] and declaring others:
freq = {} ; err = {} ; largesep = {} ; freq_p = {}
for l in xrange(l_max+1):
    freq[l] = {}
    err[l] = {}
    largesep[l] = {}
    freq_p[l] = {}
for (l,n),(f,e) in zip(l_n,f_e):
    freq[l][n] = f
    err[l][n] = e
    print ("l=%i \t n=%i \t %.2f +- %.2f" % (l,n,freq[l][n],err[l][n]))
```

### 1.2) Calculation of seismic parameters

```python
# large separations
largesep_list = []
for l,n in l_n:
    if n<n_max[l]: 
        largesep[l][n] = freq[l][n+1] - freq[l][n]
        largesep_list.append(largesep[l][n])
m_ls = {}
for l in xrange(l_max+1):
    m_ls[l] = np.mean(largesep[l].values())
    print ("m_ls[%i] = %.4f" % (l,m_ls[l]))
m_ls_T = np.mean(m_ls.values())
print 'mean Ls total:',m_ls_T
```

```python
# small separation l=0,2:
smallsep02 = {}
for n in xrange(max(n_min[0],n_min[2]+1), min(n_max[0],n_max[2]+1)+1):
    smallsep02[n] = freq[0][n] - freq[2][n-1]
    print n, smallsep02[n]
m_ss02 = np.mean(smallsep02.values())
print 'mean ss02:',m_ss02
```

```python
# r02
r02 = {}
for n in smallsep02:
    if n in largesep[1]:
        r02[n] = smallsep02[n] / largesep[1][n]
m_r02 = np.mean(r02.values())
print 'mean r02:',m_r02
```

```python
# r01
#r01 = {}
r01 = [] ; f_r01 = []
for n in xrange(max(n_min[0]+1,n_min[1]+1), min(n_max[0]-1,n_max[1])+1):
    #r01[n] = (freq[0][n-1] - 4*freq[1][n-1] + 6*freq[0][n] - 4*freq[1][n] + freq[0][n+1]) / (8*(freq[1][n]-freq[1][n-1]))
    #print r01[n]
    f_r01.append(freq[0][n])
    r01.append((freq[0][n-1] - 4*freq[1][n-1] + 6*freq[0][n] - 4*freq[1][n] + freq[0][n+1]) / (8*(freq[1][n]-freq[1][n-1])))
#print r01
m_r01 = np.mean(r01) # .values())
print 'mean r01:',m_r01
```

```python
# r10
#r10 = {}
r10 = [] ; f_r10 = []
for n in xrange(max(n_min[0],n_min[1]+1), min(n_max[0]-1,n_max[1]-1)+1):
    #r10[n] = -(freq[1][n-1] - 4*freq[0][n] + 6*freq[1][n] - 4*freq[0][n+1] + freq[1][n+1]) / (8*(freq[0][n+1]-freq[0][n]))
    #print r10[n]
    f_r10.append(freq[1][n])
    r10.append(-(freq[1][n-1] - 4*freq[0][n] + 6*freq[1][n] - 4*freq[0][n+1] + freq[1][n+1]) / (8*(freq[0][n+1]-freq[0][n])))
#print r10
m_r10 = np.mean(r10) #.values())
print 'm_r10:',m_r10
```

```python
# r010
r010 = [] ; f_r010 = []
for i in xrange(len(r10)):
    f_r010.append(f_r01[i])
    r010.append(r01[i])
    f_r010.append(f_r10[i])
    r010.append(r10[i])
#print f_r010, r010
# average and slope in particular range:
frmin = 1450 ; frmax = 1900
f_r010_r = [f_r010[i] for i in xrange(len(r010)) if f_r010[i]>frmin and f_r010[i]<frmax]
r010_r = [r010[i] for i in xrange(len(r010)) if f_r010[i]>frmin and f_r010[i]<frmax]
#print f_r010_r, r010_r
m_r010_r = np.mean(r010_r)
print 'm_r010_r = ',m_r010_r
# slope
#sl_r010_r = np.mean(np.diff(r010_r, n=1) / np.diff(f_r010_r, n=1))
sl_r010_r = stats.linregress(f_r010_r,r010_r)[0]
print 'sl_r010_r = ',sl_r010_r
```

### 1.3) Start perturbations
We perturb the observed frequencies with gaussian noise (std of the distribution is the 1-sigma observed error)

```python
numpert = 3.e+6
#np.random.seed(seed=1313)
for (l,n),(f,e) in zip(l_n,f_e):
    freq_p[l][n] = np.random.normal(f, e, numpert)
#print freq[0][13], err[0][13], freq_p[0][13]

# large separations for all:
for l,n in l_n:
    if n<n_max[l]: 
        largesep[l][n] = freq_p[l][n+1] - freq_p[l][n]
for l in xrange(l_max+1):
    m_ls[l] = np.mean(largesep[l].values(), axis=0)
    #print 'm_ls[',l,'] = ',m_ls[l]
m_ls_T = np.mean(m_ls.values(), axis=0)
#print 'mean Ls_total for all perturbations:',m_ls_T

# small separations l=0,2 for all:
smallsep02 = {}
for n in xrange(max(n_min[0],n_min[2]+1), min(n_max[0],n_max[2]+2)):
    smallsep02[n] = freq_p[0][n] - freq_p[2][n-1]
m_ss02 = np.mean(smallsep02.values(), axis=0)
#print 'mean ss02 for all perturbations:',m_ss02

# r02 for all
r02 = {}
for n in smallsep02:
    if n in largesep[1]:
        r02[n] = smallsep02[n] / largesep[1][n]
m_r02 = np.mean(r02.values(), axis=0)
#print 'mean r02:',m_r02

# r01 for all
r01 = [] ; f_r01 = []
for n in xrange(max(n_min[0]+1,n_min[1]+1), min(n_max[0]-1,n_max[1])+1):
    f_r01.append(freq_p[0][n])
    r01.append((freq_p[0][n-1] - 4*freq_p[1][n-1] + 6*freq_p[0][n] - 4*freq_p[1][n] + freq_p[0][n+1]) /
               (8*(freq_p[1][n]-freq_p[1][n-1])))
m_r01 = np.mean(r01, axis=0)
#print 'm_r01:',m_r01

# r10 for all:
r10 = [] ; f_r10 = []
for n in xrange(max(n_min[0],n_min[1]+1), min(n_max[0]-1,n_max[1]-1)+1):
    f_r10.append(freq_p[1][n])
    r10.append(-(freq_p[1][n-1] - 4*freq_p[0][n] + 6*freq_p[1][n] - 4*freq_p[0][n+1] + freq_p[1][n+1])
               / (8*(freq_p[0][n+1]-freq_p[0][n])))
m_r10 = np.mean(r10, axis=0)
#print 'm_r10:',m_r10

# r010 for all:
r010 = [] ; f_r010 = []
for i in xrange(len(r10)):
    f_r010.append(f_r01[i])
    r010.append(r01[i])
    f_r010.append(f_r10[i])
    r010.append(r10[i])
#print f_r010, r010

# average and slope in particular range:
frmin = 1450 ; frmax = 1900
f_r010_r = [f_r010[i] for i in xrange(len(r010)) if f_r010[i][0]>frmin and f_r010[i][0]<frmax]
r010_r = [r010[i] for i in xrange(len(r010)) if f_r010[i][0]>frmin and f_r010[i][0]<frmax]
#print f_r010_r
#print r010_r
m_r010_r = np.mean(r010_r, axis=0)
#print 'm_r010_r = ',m_r010_r
sl_r010_r = []
for f,r in zip(f_r010_r,r010_r):
    #print f,r,stats.linregress(f,r)[0]
    sl_r010_r.append(stats.linregress(f,r)[0])
#print sl_r010_r
```

### 1.4) Covariance matrix

```python
print '** testing the error for a simple large and small separations (no mean of several)'
print (np.std(largesep[0][16]))**2
print np.corrcoef(largesep[0][14], smallsep02[20])
cov = np.cov(largesep[0][14], smallsep02[20])
print cov
print 'the error is correct (diag is sig2):',np.sqrt(np.diag(cov))
```

```python
print '** CORRELATIONS between m_ls_T, m_ss02, m_r010_r, sl_r010_r **' # m_r02
mat_all = zip(m_ls_T, m_ss02, m_r010_r, sl_r010_r) # m_r02
print 'normalized:\n', np.corrcoef(mat_all, rowvar=0)
cov = np.cov(mat_all, rowvar=0)
print 'covariance matrix:',cov
print 'errors(std) in each param (sqrt of diag(cov)):',np.sqrt(np.diag(cov))
```
